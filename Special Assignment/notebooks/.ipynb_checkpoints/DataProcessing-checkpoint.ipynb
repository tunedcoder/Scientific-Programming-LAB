### Processing Time Dependent (Periodic) Features 
Parsing Object Type of Date,Time

```python
def Time_parser(df):
    df["Year"] = df["Date"].str.split("-").str[0]
    df["Month"] = df["Date"].str.split("-").str[1]
    df["Date_split"] = df["Date"].str.split("-").str[2]
    df["sunrise h"]= df["sunrise"].str.split(" ").str[1].str.split(":").str[0]
    df["sunrise m"]= df["sunrise"].str.split(" ").str[1].str.split(":").str[1]
    df["sunrise s"]= df["sunrise"].str.split(" ").str[1].str.split(":").str[2]
    df["sunset h"]= df["sunset"].str.split(" ").str[1].str.split(":").str[0]
    df["sunset m"]= df["sunset"].str.split(" ").str[1].str.split(":").str[1]
    df["sunset s"]= df["sunset"].str.split(" ").str[1].str.split(":").str[2]
    
    df = df.drop(columns=["Date","sunrise","sunset"])
    return df
```

```python
df = Time_parser(df)
```

```python
df
```

### Encoding All Output Ordinally and Type Cating Transient Features

```python
df_new = df.dropna()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
Encoders ={}
```

```python
df_new.select_dtypes("object")
```

```python
o_list = df.select_dtypes("object").dtypes[2:].index.tolist()
o_list
df_new_ew = df_new.copy()
df_new_ew[o_list] = df_new[o_list].astype(str).astype(int)
df_new = df_new_ew
```

```python
df_new.reset_index(inplace=True,drop=True)
```

```python
df_new
```

```python
Encoders["City"] = OrdinalEncoder()
Encoders["conditions"] = OrdinalEncoder()
```

```python
df_new["City"] = Encoders["City"].fit_transform(np.array(df_new["City"]).reshape(-1,1))
df_new["conditions"] = Encoders["conditions"].fit_transform(np.array(df_new["City"]).reshape(-1,1))
```
