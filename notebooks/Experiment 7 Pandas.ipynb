```python
import numpy as np
import pandas as pd
```

```python
device_df = pd.read_csv("../Data/user_device.csv")
usage_df = pd.read_csv("../Data/user_usage.csv")
```

```python
device_df
```

```python
usage_df
```

```python
data = device_df.merge(usage_df,on="use_id",how="left")
```

```python
data
```

```python
data = data.dropna()
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data
```

```python
data.device.unique()
```

```python
data["Is Samsung"] = data.device.str.startswith("SM")
data["Is Apple"] = data.device.str.startswith("iP")
data["Is LG"] = data.device.str.startswith("LG")
data["Is HTC"] = data.device.str.startswith("HTC")
```

```python
data[data["Is HTC"]]
```

```python
data[data["Is HTC"]]["outgoing_mins_per_month"].hist()
```

```python
data[data["Is Samsung"]]["outgoing_mins_per_month"].sum()
```

```python
print(f'Average Of HTC Mins {data[data["Is HTC"]]["outgoing_mins_per_month"].mean()}')
print(f'Average Of Samsung Mins {data[data["Is Samsung"]]["outgoing_mins_per_month"].mean()}')
```
