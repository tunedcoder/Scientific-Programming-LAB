```python
import matplotlib.pyplot as plt
import math
import itertools
import numpy as np

%matplotlib inline
```

```python
def bellcurve(x,s,m):
    a = []
    for e in x:
        a.append((math.exp(-0.5 * ((e-m)/s)**2))/(s*(2*math.pi)**2))
    return np.array(a)

def bellcurve_custom(x_raw):
    m = np.array(x_raw).mean()
    s = np.array(x_raw).var()
    
    return bellcurve(x_raw,s,m)

def kmph_to_mps(a):
    return a/3.6

def stopping_distance(v_o,mu):
    g = 9.8
    return ((kmph_to_mps(v_o)**2)/(mu*g))/2
def stopping_custom(v):
    return [stopping_distance(e,0.3) for e in v]
```

```python
m = float(input("Enter Mean: "))
s = float(input("Enter Variance (S): "))
x = np.linspace(m-5*s,m + 5*s,100)
y = bellcurve(x,s,m)
plt.plot(x,y)
```

```python
x_raw = eval(input("Enter comma Seprated Values:"))
```

```python
s = list(x_raw)
s.sort()
y_custom = bellcurve_custom(s)
plt.plot(s,y_custom)
```

```python
start_vel = int(input("Enter Start Velocity: "))
end_vel = int(input("Enter End Velocity: "))
ss = float(input("Enter Step_Size: "))
```

```python
x_vel = np.linspace(start_vel,end_vel,int((end_vel-start_vel)/ss))
```

```python
dis=stopping_custom(x_vel)
```

```python
plt.plot(x_vel,dis)
```
